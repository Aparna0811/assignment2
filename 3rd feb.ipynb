```python
# 1. def keyword is used for create to a function.

def odd_num(x):
    return [ x for x in range(x+1) if x % 2 !=0 ]
print(odd_num(25))
```

```python
# *args allow to pass multiple input in function
def test1(*args):
    return args
test1(1,2,"aparna",[40,-5])
```

```python
# **kwargs return dictionary value
def test2(**kwargs):
    return kwargs
test2(a=[1,2,3],b="hello",c=233.45)
```

```python
n=int(input("enter the number"))
temp=n
rev=0
while (n>0):
    d=n%10
    rev=rev*10+d
    n=n//10
if (temp==rev):
    print("the number is palindrome")
else:
    print("the number is not palindrone")
```

```python
# generator function returns an iterator that produces a sequence of values when iterated over
# yield keyword is used to control the flow of a generator function.similar to return statement
def count(n):
    count=1
    while count<n:
        yield count
        count=count+1
c=count(5)
```
